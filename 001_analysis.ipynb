## Spatial Stair Scaffolding

As the further a stair is from the bottom of the staircase, the more the rise of the stair (relative to the previous stair). Temperature controls whether the staircase starts flat and only rises at the end (temp <= -1), whether the relative rise is linearly increasing (temperature = 0) or whether the staircase is maximally steep (temp >= 1)

```python
xs = np.linspace(0, 1, 101) # stair position, from 0 .. 1.
temps = np.linspace(-1, 1, 5) # scaffolding temperature, roughly -1 to 1.
temps = np.array([-1, -0.25, 0, 0.25, 1]) # possible scaffolding schedule
for temp in temps: 
    if temp >= 0:
        zs = 1 - np.power(1 - xs, np.exp(4 * temp))
    else:
        zs = np.power(xs, np.exp(-4 * temp))
        
#     print(f'temp: {temp:0.2}')
#     print(*[f'{z:0.4}' for z in zs[:10]])
#     print(*[f'{z:0.4}' for z in zs[10:]])
#     plt.figure(figsize=(6,6))
    plt.plot(xs, zs, label=f't = {temp:.2}')
#     plt.legend()
    
    plt.xlabel('Stair Position')
    plt.ylabel('Stair Rise')
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.title(f'Stair Rise vs Position by Temperature (t={temp:.2})')
    plt.axis('scaled')
    plt.show()
    print('===============')
```

```python
# Inspired by squircle https://en.wikipedia.org/wiki/Squircle
xs = np.linspace(0, 1, 101)
ps = np.linspace(1, 4, 5)
for p in ps:
    ys = 1 - np.power(1 - np.power(xs, p), 1/p)
    plt.plot(xs, ys, label=f'p={p}')
for p in ps:
    ys = np.power(1 - np.power(1 - xs, p), 1/p)
    plt.plot(xs, ys, label=f'p={p:.4}')
    
plt.axis('equal')
plt.legend()
plt.show()
```
